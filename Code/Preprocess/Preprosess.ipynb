```python
import pandas as pd
from tqdm import tqdm
import os
from langdetect import detect
from underthesea import sent_tokenize, text_normalize, word_tokenize
```

# II. Xây dụng hàm cần thiết

## 1. Kiểm tra xem phải tiếng việt hay không

```python
def is_vietnamese(sentence):
    try:
        # Phát hiện ngôn ngữ của câu
        language = detect(sentence)
    except:
        return False

    # Kiểm tra xem ngôn ngữ có phải là tiếng Việt hay không
    if language == 'vi':
        return True
    else:
        return False

```

```python
text_1 = 'Đội ngũ bảo trì quá thưa thớt dẫn đến không đảm bảo được chất lượng sửa chữa thiết bị.'
text_2 = 'The maintenance team is too sparse, making it impossible to guarantee the quality of equipment maintenance.'

if is_vietnamese(text_1):
  print('"text_1" là tiếng việt')
else:
  print('"Text_1" không là tiếng việt')

if is_vietnamese(text_2):
  print('"Text_2" là tiếng việt')
else:
  print('"Text_2" không là tiếng việt')
```

```python
text = 'Đội ngũ bảo trì quá thưa thớt dẫn đến không đảm bảo được chất lượng sửa chữa thiết bị.'
print(word_tokenize(text, format="text"))
```

# III. Xử lí dữ liệu

## 1. Vietnamese Students’ Feedback Corpus (UIT-VSFC)

### 1.2. Trộn ngẫu nhiên

```python
df = df.sample(frac=1).reset_index(drop=True)
df
```

### 1.3. Xử lí

```python
for file_name in os.listdir(base_dir):
    duong_dan_tuyet_doi = os.path.join(base_dir, file_name)
    if os.path.isfile(duong_dan_tuyet_doi):
        print(f'Đang xử lý file: {file_name}')
        # Xử lý file ở đây
        df = pd.read_csv(duong_dan_tuyet_doi)
        df
        # Xoá hàng bị thiếu dữ liệu
        df.dropna()
        df.shape
        # Xoá hàng bị trùng câu nhận xét
        df.drop_duplicates(subset='sentence', inplace=True)
        df.shape
        tqdm.pandas()
        # Xoá những bình luận không phải tiếng việt
        print('\tĐang xoá bình luận không phải tiếng việt:')
        df['is_vietnamese'] = df['sentence'].progress_apply(is_vietnamese)
        df = df[df['is_vietnamese']].copy()
        df.drop(columns=['is_vietnamese'], inplace=True)
        df.shape
        # Tokenize
        print('\tĐang tokenize từ:')
        df['sentence'] = df['sentence'].progress_apply(lambda x: word_tokenize(x, format="text"))

        # Đường dẫn tới thư mục
        preprocess_path = base_dir+'preprocess'

        # Kiểm tra xem thư mục có tồn tại hay không
        if not os.path.exists(preprocess_path):
            # Nếu không, tạo thư mục
            os.makedirs(preprocess_path)

        # Lưu DataFrame vào file CSV trong thư mục
        df.to_csv(os.path.join(preprocess_path, 'preprocess_'+str(file_name)), index=False)

```
